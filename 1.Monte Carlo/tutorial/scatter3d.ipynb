```python
import numpy as np
import matplotlib.pyplot as plt

###  ######
t = np.linspace(0, np.pi * 8, 500)
x = np.cos(t)
y = np.sin(t)
z = t
###################################

###  #######################
# 
fig = plt.figure(figsize=(10,10), dpi=100)
# 
ax = plt.axes(projection='3d')
# 
ax.scatter3D(x, y, z)
##########################################
```

```python
import numpy as np
import matplotlib.pyplot as plt

### 这是一条螺旋线上的一些点 ######
t = np.linspace(0, np.pi * 8, 500)
x = np.cos(t)
y = np.sin(t)
z = t
###################################

### 绘制三维散点图 #######################
# 设置画布大小
fig = plt.figure(figsize=(10,10), dpi=100)
# 创建三维图
ax = plt.axes(projection='3d')
# 绘制三维散点图
ax.scatter3D(x, y, z)
##########################################
```

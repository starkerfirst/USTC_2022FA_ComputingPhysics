```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

r = 1 # 半径
a, b = (0.0, 0.0) # 圆心

# 正方形边界
xmin, xmax = a-1, a+1
ymin, ymax = b-1, b+1

# 随机抽取n个点
n = 100000

# 在正方形范围内，随机抽取n个点
x = np.random.uniform(xmin, xmax, n)
y = np.random.uniform(ymin, ymax, n)

# 计算每个点到圆心的距离
d = np.sqrt((x-a)**2 + (y-b)**2)

# 园内点的个数
res = sum(np.where(d<r, 1, 0)) # 距离小于半径即为圆内

# pi
pi = res/n*4
print(pi)
```

```python
# 绘制正方形
fig = plt.figure(figsize=(10, 10))
axes = fig.add_subplot(1, 1, 1)
plt.plot(x, y, 'ro', markersize=0.2)


# 绘制圆形
circle = Circle(xy=(a, b), radius=r, alpha=0.5, color='gray')
axes.add_patch(circle)
plt.grid(True, linestyle='--', linewidth=0.8)
```

Numpy中提供了Ndarray对象用来表示高维数组  
用np.array方法可以将列表或元组等对象转换成Ndarray

```python
np.array((1, 2, 3))
```

```python
np.array(([1, 2, 4], [2, 3, 1]))
```

Ndarray的维数不限，一维和二维比较常用

用Ndarray来计算可以非常方便地同时修改数组中的每一个数

```python
a = np.array((1,2,3))
a + 1
```

```python
a * 2
```

```python
b = np.array((4,1,3))
a + b
```

在二维的情况下可以做一些矩阵操作,例如转置等

```python
a = np.array(((1,2,3),(4,5,6)))
print(a)
print(a.T)
```

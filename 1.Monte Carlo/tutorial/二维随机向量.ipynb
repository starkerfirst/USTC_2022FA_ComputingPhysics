用舍选法产生二维随机向量

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def fxy(x, y):
    return (x + y)/2
X = []
Y = []
N = 0
#start = time.time()
while N < 10000:
    a, b, c = np.random.random(3)
    if fxy(a, b) >= c:
        X.append(a)
        Y.append(b)
        N += 1
#print(f"Time: {time.time() - start} s")
```

```python
plt.figure(figsize=(10,10), dpi=100)
plt.scatter(X, Y, 5)
plt.show()
```

```python
# 绘制三维柱形图的准备工作
H, x_, y_ = np.histogram2d(X, Y, bins=10, range=[[0,1],[0,1]])
x_ = (x_[1:] + x_[:-1])/2
y_ = (y_[1:] + y_[:-1])/2
xx_, yy_ = np.meshgrid(x_, y_)
x_, y_ = xx_.ravel(), yy_.ravel()
top = H.ravel() / 100.0
bottom = np.zeros_like(top)
width = depth = 0.1
```

```python
# 绘制三维网格图的准备工作
x = np.linspace(0,1,11)
y = np.linspace(0,1,11)
x, y = np.meshgrid(x, y)
z = fxy(x,y)*2
```

```python
fig = plt.figure(figsize=(10, 10), dpi=100)
ax = plt.axes(projection='3d')
ax.bar3d(x_, y_, bottom, width, depth, top, shade=True, color='yellow', label='XY Distribution')
ax.plot_wireframe(x, y, z, color='blue', label='Theoretical Distibution')
plt.show()
```

```python
import datetime

#创建一个列表来存储随机数
randomlist = []
 
#设置一个数字的随机数作为种子
x0 = 6406

#以randomlist的id后4位为随机数种子
#seed = id(randomlist) % 10000

#以系统时间毫秒数的后4位为随机数种子
#seed = datetime.datetime.now().microsecond % 10000

a = 24298
c = 99991
m = 199617

for i in range(0,2000):
    x0 = (x0*a+c)%m
    randomlist.append(x0)
    #print(randomlist[i])
```

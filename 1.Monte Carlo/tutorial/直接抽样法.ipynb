使用直接抽样法对概率密度函数如下的随机变量进行抽样$$f(x)=2x, x\in[0,1]$$
对概率密度函数积分得到分布函数$$F(x)=x^2, x\in[0,1]$$
令$\xi=F(\eta)=\eta^2$，反解出$\eta=\sqrt{\xi}$  
对$[0,1]$范围内均匀分布的随机变量$\xi$做以上变换，得到的$\eta$就是服从目标分布的随机变量

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 产生[0,1]范围内均匀分布的随机变量xi
xi = np.random.random(10000)
# 通过变换得到eta
eta = np.sqrt(xi)
```

```python
# 理论密度函数曲线
x = np.linspace(0.0, 1.0, 11)
y = 2 * x
```

```python
# 比较理论分布和抽样所得的实际分布
plt.figure(figsize=(5, 5), dpi=100)
plt.plot(x, y, color='blue', label='Theoretical Distribution')
plt.hist(eta, bins=10, range=(0.0,1.0), density=True, color='yellow', label='Random variable X distribution')
plt.legend(loc='upper left')
plt.show()
```
